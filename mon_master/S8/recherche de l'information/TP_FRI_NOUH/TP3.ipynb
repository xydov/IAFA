````
def outil(text, model, index):
    res = []
    mots = text.split()
    mots_majuscules = [mot for mot in mots if mot[0].isupper()]

    for mot in mots_majuscules:
        # Au lieu d'imprimer les r√©sultats, les stocker dans une liste
        results = pt.BatchRetrieve(index, wmodel=model, metadata=["docno", "title", "url"]).search(mot).head(10)
        res.append(results)

    return res
````

````
def dic(requetes, model, index):
    # Initialisation des variables
    size = 100
    cpt = 101   # Num√©ro de d√©part pour les ensembles de documents
    dics = {}   # Dictionnaire final pour stocker les r√©sultats

    # Parcours de chaque requ√™te dans la liste des requ√™tes
    for r in requetes:
        # Appel de la fonction outil pour obtenir les documents pertinents pour la requ√™te
        docs = outil(r, model, index)

        # Parcours de chaque document obtenu pour la requ√™te
        for i, doc in enumerate(docs):
            # Calcul du num√©ro unique pour chaque ensemble de documents
            num = cpt + i

            # Cr√©ation d'un dictionnaire pour stocker les r√©sultats du document actuel
            doc_dict = {}

            # Parcours des titres et des scores de chaque document et ajout au dictionnaire
            for title, score in zip(doc['title'], doc['score']):
                # Remplacement des espaces par des underscores dans les titres pour √©viter les espaces dans les cl√©s du dictionnaire
                doc_dict[title.replace(" ", "_")] = score

            # Ajout du dictionnaire de r√©sultats du document au dictionnaire global avec le num√©ro unique comme cl√©
            dics[str(num)] = doc_dict

        # Mise √† jour du num√©ro pour le prochain ensemble de documents
        cpt += size

    # Retour du dictionnaire global contenant tous les ensembles de documents pour chaque requ√™te
    return dics
````

Avec les m√™mes 5 configurations, g√©n√©rez 1000 r√©sultats et √©valuez-les. Il y a-t-il des diff√©rences dans certains m√©triques ? Pourquoi¬†?


````
def dic2(requetes,model,index):
  size = 1000 #1000 r√©sultats √† g√©nerer
  cpt = 101
  dics = {}
  for r in requetes :
    docs = outil(r, model, index)
    for d in range(len(docs)) :
      dic = {}
      num = cpt + (d)
      for t in range(len(docs[d]['title'])):
        dic[docs[d]['title'][t].replace(" ","_")] = docs[d]['score'][t]
      dics[str(num)] = dic

    cpt = 100 + cpt
  return dics
````
