```python
import numpy as np
import matplotlib.pyplot  as plt
img = plt.imread('2cvNG.jpg')
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img = plt.imread('2cvBRUIT.jpg')
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img_debruit = np.copy(img)
taille=np.shape(img_debruit)

for x in range(1, taille[0]-1):
    for y in range(2, taille[1]-1):
        v = np.sort(np.ravel(img_debruit[x-1:x+2, y-1:y+2]))
        img_debruit[x-1, y-1] = v[4]

# Affichage
plt.figure(5)
plt.imshow(img_debruit, cmap='Greys_r')
plt.title('Image débruitée ')
plt.savefig("debruit.jpg")
plt.show()
```
