**Question TP5.1 (2pts)**: Effectuez un rehaussement Laplacien de l'image ci-dessous.



**Réponse**: *(mettez le code source du (des) programme(s) dans le box ci-dessous)*

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
img = plt.imread('2cv_float.jpg')
plt.figure(7)
plt.imshow(img)
plt.draw()
masque_Laplacien = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
img_Laplacien = np.zeros(img.shape)
img_Laplacien[:,:,0] = convolve2d(img[:,:,0], masque_Laplacien, mode='same')
img_Laplacien[:,:,1] = convolve2d(img[:,:,1], masque_Laplacien, mode='same')
img_Laplacien[:,:,2] = convolve2d(img[:,:,2], masque_Laplacien, mode='same')
img_Laplacien_up = img - img_Laplacien
img_Laplacien_up = np.uint8(np.clip(img_Laplacien_up, 0, 255))

plt.figure(8)
plt.imshow(img_Laplacien_up)
plt.show()
```

**Question TP5.2**: a) (1pt) Trouvez une valeur de seuil sur la composante Rouge qui permette de garder uniquement le vert (partie gauche) du drapeau italien ci-dessous.




**Réponse**: *(mettez le code source du (des) programme(s) dans le box ci-dessous)*

```python
img = plt.imread('/content/drapeau_Italie.jpg')

plt.figure(10)
plt.imshow(img)
plt.draw()
dimension = np.shape(img)
s = 100
masque = (img[:,:,0] < s)
img_g = np.zeros(dimension, dtype=np.uint8)
img_g[masque, :] = img[masque, :]
plt.figure(10)
plt.imshow(img_g)
plt.draw()
```
