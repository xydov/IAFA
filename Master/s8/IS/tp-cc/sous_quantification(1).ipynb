```python
import matplotlib.pyplot  as plt
import matplotlib.pyplot as plt
import numpy as np
def sousquantifierimage(input,bits):
  img = plt.imread(input)
  decalage = 8 - bits #pour garder les bits
  mask = np.uint8(2**bits - 1) << decalage
  imageSq = np.bitwise_and(img, mask)
  return imageSq

img2=sousquantifierimage('/content/crayons(1).jpg',2)
plt.imshow(img2)
plt.show()
plt.savefig('crayons-2bits.jpg')
```
