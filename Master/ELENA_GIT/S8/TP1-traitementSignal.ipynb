```python
import scipy.io.wavfile as wav
fs, signal = wav.read('diner.wav')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#frequence d'echantillonage
print("fs =",fs)
#transformer d'entiers a reels
floatSignal = signal / 32767.0
print(floatSignal)
#calcul de la duree du fichier
duree = floatSignal.size / fs
print("duree =",duree)
#afficher le signal
temps = np.arange(0, duree,(1/fs))
plt.plot(temps,floatSignal)
plt.xlabel("Temps en secondes")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Affichez les 10 premires valeurs du signal qui sont diffrentes de 0
non_zero = floatSignal[np.nonzero(floatSignal)]
premiers_non_zeros = non_zero[:10]
print(premiers_non_zeros)
```

```python
from scipy.io import wavfile
# fonction de sous-quantification un fichier audio
def sousquantifier(input, output,bits):
    fs, audio = wavfile.read(input)
    audio = audio / np.max(np.abs(audio))
    quantization_levels = 2 ** bits
    sous_quantifier= np.round((audio + 1) / 2 * (quantization_levels - 1)) / (quantization_levels - 1) * 2 - 1
    wavfile.write(output, fs, sous_quantifier)
```

```python
sousquantifier('diner.wav','sousquantifier12.wav',12)
sousquantifier('diner.wav','sousquantifier8.wav',8)
sousquantifier('diner.wav','sousquantifier4.wav',4)
sousquantifier('diner.wav','sousquantifier2.wav',2)
#On remarque plus le nombre de bits est petit plus la qualite de l'audio diminue ainsi que l'apparition d'un bruit
```

```python
# fonction de sous-echantillonage
def sousechantilloner(input, output, nb):
    fs, audio = wavfile.read(input)
    downsampled = audio[::nb]
    wavfile.write(output, fs // nb, downsampled)
```

```python
sousechantilloner('diner.wav','sousechantilloner2.wav',2)
sousechantilloner('diner.wav','sousechantilloner4.wav',4)
sousechantilloner('diner.wav','sousechantilloner8.wav',8)
#On remarque que plus on sous-echantionne l'audio plus on a une perte de qualite et la reduction de la taille du fichier
```

```python
# fonction de sous-quantification image
def sousquantifierimage(input,bits):
  img = plt.imread(input)
  decalage = 8 - bits #pour garder les bits 
  mask = np.uint8(2**bits - 1) << decalage
  imageSq = np.bitwise_and(img, mask)
  return imageSq
```

```python
img2=sousquantifierimage('photo.jpg',2)
img3=sousquantifierimage('photo.jpg',3)
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
```

On observe une reduction de qualite d'image ainsi que l'apparition des bandes de couleurs qui implique une perte de precision.
